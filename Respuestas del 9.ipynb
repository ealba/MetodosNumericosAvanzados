# El método de Newton

#### Recordemos que el *método de Newton* es un método numérico *iterativo* para encontrar raíces de funciones (continuamente) diferenciables.  Sea $f$ la función cuyas raíces queremos encontrar. La idea es que empecemos desde una adivinanza inicial $x_0$, y que la siguiente aproximación $x_1$ esté donde la recta tangente a la curva $f(x)$ en $x_0$ corte el eje $x$.   [1] Esboza la geometría, tanto a mano como en la computadora, y así encuentra la expresión de $x_{n+1}$ en términos de $x_n$.

Bien ilustrado en:

http://commons.wikimedia.org/wiki/File:NewtonIteration_Ani.gif

Para el primer paso de la iteración se toma $x_0$ aleatoriamente y después (siempre que no se tenga un punto crítico):

### $x_{n+1}=x_n - \frac{f(x_{n})}{f'(x_{n})}$

### [2] Implementa la iteración para calcular la raíz cuadrada y la raíz cúbica de $2$.  ¿Cuál es una condición razonable de terminación del algoritmo?

Para sacar la raíz cuadrada de 2, resolvemos la ecuación: $x^2-2=0$. De igual manera, para obtener la raíz cúbica de 2 resolvemos:  $x^3-2=0$.

Es razonable terminar el algoritmo cuando :

a) Se excede un número límite de iteraciones

b) Se alcance una tolerancia dada por el usuario. En este caso podemos decir que el algoritmo alcanza su meta cuando: $\frac{x_n - x_{n-1}} {x_n} \leq TOLERANCIA$

```julia
using Diff
```

```julia
using Intervalos
```

En esta versión se tiene que dar la derivada implícita como f1:

```julia
function NewtonRhapsod (x0,iteraciones::Integer,tol,f::Function,f1::Function)
for i in 1:iteraciones
    raiz=x0-f(x0)/f1(x0)
    
    if (abs(raiz-x0) < tol*abs(raiz))
        iteraciones=i    
        break
    
    else
        x0=raiz
    end
end
#println("la raíz es"","" ""después de"" ""$iteraciones"" ""iteraciones") 
return x0
end
```

```julia
f(x)= x^2-2.0;
iteraciones=5015;
x0=1.0;
f1(x)=2x;
tol=2.0^-30;

@time NewtonRhapsod(x0,iteraciones,tol,f,f1)
```

De las gráficas anteriores vemos que hay una bisectriz que divide (en general) el tipo de raíces. Cabe mencionar que el origen es un punto crítico de la derivada de esta función y por eso se tiene que el método de Newton Rhapsod explota allí.

caso 1) Cuando se toma un punto inicial con parte imaginaria cero, o por lo menos con el valor absoluto de la parte real mucho mayor que la de la parte imaginaria, el resultado arroja (siempre que el punto inicial no sea el origen) la raíz $r_1=1$

caso 2) Cuando se toma un punto inicial con parte imaginaria negativa comparable, en valor absoluto, con la parte real, el resultado arroja la raíz $r_2=  \frac{-1}{2}-\frac{i \sqrt3}{2}$

caso 3) Cuando se toma un punto con parte imaginaria positiva comparable, en valor absoluto, con la parte real, el resultado arroja la raíz $r_3=  \frac{-1}{2}+\frac{i \sqrt3}{2}$

### [5] Desarrolla e implementa el método de Newton para funciones $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^n$. Para hacerlo, toma una adivinaza $\mathbf{x}_n$ y resuelve la ecuación $\mathbf{f}(\mathbf{x}_{n+1}) = \mathbf{0}$, con $\mathbf{x}_{n+1} = \mathbf{x}_n + \delta \mathbf{x}_n$.

El algoritmo en este caso se debe cambiar usando $\delta x$ = $-J^{-1}_{f}(\mathbf x) f(\mathbf x)$. Donde $J^{-1}_{f}$ es la matríz jacobiana inversa

Tomamos por ejemplo: Cuando n=2

### [6] Utiliza tu algoritmo para calcular raíces de funciones conocidas multidimensionales.

```julia
function NewtonRhapsod2DIMENSIONAL(x0,iteraciones,tol,h::Function)
    #---------las derivadas parciales casi a mano (¿se puede hacer automático?)
h1x(x,y)=h(DV(x,1),DV(y,0))[1].d
h2x(x,y)=h(DV(x,1),DV(y,0))[2].d
h1y(x,y)=h(DV(x,0),DV(y,1))[1].d
h2y(x,y)=h(DV(x,0),DV(y,1))[2].d
    #-------------------------------------------------------------------------
    
    #---------------------------------------------
    J(x,y)=[h1x(x,y) h1y(x,y); h2x(x,y) h2y(x,y)];
    Jinv(x,y)=inv(J(x,y));
    δx(x,y)=Jinv(x,y)*[h(x,y)[1],h(x,y)[2]]
    #---------------------------------------------
    
    
    
for i in 1:iteraciones
raiz=[x0[1],x0[2]]-δx(x0[1],x0[2])
    
    if (norm(raiz-[x0[1],x0[2]]) < tol*norm(raiz))
        iteraciones=i    
        break
    
    else
        x0=raiz
    end

end
    
#println("la raíz es"","" ""después de"" ""$iteraciones"" ""iteraciones") 
return x0
    
end
```

```julia
x0=(1.7,0.5)
iteraciones=50;
tol=2.0^-50.0;
h(x,y)=(x^2+y^2-1, x^3+y^3-1)
NewtonRhapsod2DIMENSIONAL(x0,iteraciones,tol,h)
```
