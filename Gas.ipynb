```julia
type partícula
    x
    v
    m
end




type Gas
    N::Int
    partículas::Array{partícula,1}
end

function Gas(partículas::Array{partícula,1})
    Gas(length(partículas), partículas)
end
```

# ------------------Función para generar  N partículas----------------------

```julia
function NGas (N::Int)
partículasazar = partícula[]  
x=rand(N,2)*((-1.0)^rand(0:1));    #Posiciones aleatorias
v=rand(N,2)*10*((-1.0)^rand(0:1)); #Velocidades tienen que ser altas y aleatorias


for i in 1:N
    #push!( partículasazar, partícula(r[i,:],v[i,:],1.0) ) #posiciones y velocidades aleatorias, masa unitaria
    push!( partículasazar, partícula([x[i,1],x[i,2]],[v[i,1],v[i,2]], 1.0)  )    
end
   
    return partículasazar    
end
```

# ------------------Función para mover  N partículas----------------------

```julia
function mover!(g::Gas,ΔT::Float64, a::Vector{Float64})

        for i in 1:g.N 
        g.partículas[i].x+=g.partículas[i].v*ΔT; 
        g.partículas[i].v+=a*ΔT;
        end
    return g
end
```

# --------Funciones para guardar y graficar la trayectoria de 1 partícula dadas diferentes condiciones de frontera------------

## Condiciones de frontera de colisión completamente inelástica

```julia
function seguir1_colinelás(N::Int,etiqueta::Int,L::Float64,a::Vector{Float64},T::Int, ΔT::Float64)

    #N se refiere al número de partículas en el gas
    #L es el tamaño de la caja cuadrada
    #T es el número de pasos
    #ΔT es la duración de un paso
    #a es el vector de aceleración
    #etiqueta es el nombre de la partícula que quieres seguir

Gasideal=Gas(NGas(N));
    
ymax=L;
ymin=-L; #longitudes de la caja cuadrada
xmax=L;
xmin=-L;

r=Vector{Float64}[]

    for i in 1:T   # a T tiempos
        mover!(Gasideal,ΔT,a)
        push!(r,[Gasideal.partículas[etiqueta].x[1] , Gasideal.partículas[etiqueta].x[2]])

        if r[i][1] > xmax  
            #println("choque a la derecha a t=$i")
            r[i][1]=xmax
            break
        end
        
        if r[i][1] < xmin  
            #println("choque a la izquierda t=$i")
            r[i][1]=xmin
            break
        end

 
        if r[i][2] > ymax  
            #println("choque arriba a t=$i")
            r[i][2]=ymax
            break
        end

        if r[i][2] < ymin  
            #println("choque abajo a t=$i")
            r[i][2]=ymin 
            break
        end 
      
       
    end

    return r
    
end
```

```julia
#seguir1_colinelás(N::Int,etiqueta::Int,L::Float64,a::Vector{Float64},T::Int, ΔT::Float64)
ΔT=1.0;
T=500;
r=seguir1_colinelás(10,1,500.00,[0.0 , 1.0],T ,ΔT);

using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
plot(x, y, color="red", linewidth=1.0)
title("Trayectoria de duración $(ΔT*T)")
xlabel("x")
ylabel("y")
```

## Colisiones completamente elásticas

```julia
function seguir1_colelás(N::Int,etiqueta::Int,L::Float64,a::Vector{Float64},T::Int, ΔT::Float64)

    #N se refiere al número de partículas en el gas
    #L es el tamaño de la caja cuadrada
    #T es el número de pasos
    #ΔT es la duración de un paso
    #a es el vector de aceleración
    #etiqueta es el nombre de la partícula que quieres seguir

Gasideal=Gas(NGas(N));
    
ymax=L;
ymin=-L; #longitudes de la caja
xmax=L;
xmin=-L;

r=Vector{Float64}[]

    for i in 1:T   # a T tiempos
       
        mover!(Gasideal,ΔT,a)

        #push!(x,Gasideal.partículas[1].x[1])    
        #push!(y,Gasideal.partículas[1].x[2])

            
        if abs(Gasideal.partículas[etiqueta].x[1])  >=xmax  
            #println("choque a la derecha a t=$i")
            Gasideal.partículas[etiqueta].v[1] = -Gasideal.partículas[etiqueta].v[1]
            continue
        end
        
    

 
        if abs(Gasideal.partículas[etiqueta].x[2]) >= ymax  
            #println("choque arriba a t=$i")
            Gasideal.partículas[etiqueta].v[2] = -Gasideal.partículas[etiqueta].v[2]
            continue
        end

      
    
        push!(r,[Gasideal.partículas[etiqueta].x[1] , Gasideal.partículas[etiqueta].x[2]])
        #PONER LAS CONDICIONES DE CHOQUE Y HACERLO PARA N PARTÍCULAS
    end

    return r
    
end
```

```julia
#seguir1_colelás(N::Int,etiqueta::Int,L::Float64,a::Vector{Float64},T::Int, ΔT::Float64)
ΔT=1.0;
T=5000;
r=seguir1_colelás(10,1,50.00,[0.0 , 0.0], T , ΔT);
using PyPlot
x=[r[i][1] for i in 1:length(r)];
y=[r[i][2] for i in 1:length(r)];
plot(x, y, color="red", linewidth=1.0)
title("Trayectoria de duración $(ΔT*T)")
xlabel("x")
ylabel("y")
```
